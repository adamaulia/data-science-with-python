```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
%matplotlib inline 
```

```python
data = pd.read_excel('ID_Population.xlsx')
data.head()
```

line plot

```python
plt.xlabel('year')
plt.ylabel('population')
plt.title('grafik populasi penduduk indonesia')
plt.plot(data['Year'],data['Population'])
```

scatter plot 

```python
plt.xlabel('year')
plt.ylabel('population')
plt.title('grafik populasi penduduk indonesia')
plt.scatter(data['Year'],data['Population'])
```

histogram

```python
age = np.random.randint(1,60,size=1000)
```

```python
plt.title('sebaran usia penduduk')
plt.xlabel('usia')
plt.ylabel('jumlah')
plt.hist(age,bins=5)
```
